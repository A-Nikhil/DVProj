```python
%config IPCompleter.greedy=True 

import pandas as pd  #for csv inputs 
import matplotlib.pyplot as plot
from pathlib import Path

technology = Path('D:/projects/DataVisProject/technology')
```

```python
datasets = Path('D:/projects/DataVisProject/datasets')
mobileCSV = datasets/'tech-price-index/TechnologyIndex.csv'
mobiles = pd.read_csv(mobileCSV)
mobiles = mobiles.drop('Rank', 1)
mobiles = mobiles.head(20)
```

```python
for i in mobiles.index:
    mobiles['iPhone'][i] = mobiles['iPhone'][i].replace(',', '')
    mobiles['iPhone'][i] = float(mobiles['iPhone'][i][1:])
mobiles.head()
```

```python
# Plotting mobiles
mobiles.head()
mobiles.plot.bar()
plot.bar(mobiles['Country'], mobiles['iPhone'])
plot.xlabel('Country')
plot.ylabel('iPhone Price (in $)')
plot.figure(figsize=(20,10))
plot.show()
```

```python
plot.scatter(mobiles['Country'], mobiles['iPhone'])
plot.figure(figsize=(20,10))
plot.show()

```
